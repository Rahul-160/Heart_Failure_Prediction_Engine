```python
df['cigsPerDay'] = df['cigsPerDay'].fillna(df['cigsPerDay'].mode()[0])
df['BPMeds'] = df['BPMeds'].fillna(df['BPMeds'].mode()[0])
df['totChol'] = df['totChol'].fillna(df['totChol'].mode()[0])
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())
df['heartRate'] = df['heartRate'].fillna(df['heartRate'].mode()[0])
df['glucose'] = df['glucose'].fillna(df['glucose'].mode()[0])
```

```python
X = df.iloc[:, :-1]
X
```

```python
Y = df.iloc[:, -1]
Y
```

```python
from imblearn.over_sampling import RandomOverSampler
os = RandomOverSampler()
X_os, Y_os = os.fit_resample(X, Y)
Y_os.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_os_scaled = scaler.fit_transform(X_os)
X_os_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_os_scaled, Y_os, test_size = 0.3, random_state = 1)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
kne = KNeighborsClassifier()
svc = SVC()
gau = GaussianNB()
dec = DecisionTreeClassifier()
ran = RandomForestClassifier(n_estimators = 200, random_state = 1, criterion = 'entropy', max_depth = 5)
ada = AdaBoostClassifier(random_state = 1)
xgb = XGBClassifier(max_depth = 25, random_state = 1, base_score = 0.8)
model = [kne, svc, gau, dec, ran, ada, xgb]
for model in model:
  print(model)
  model.fit(X_train, Y_train)
  Y_pred = model.predict(X_test)
  print(classification_report(Y_test, Y_pred))
  print(ConfusionMatrixDisplay.from_predictions(Y_test, Y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
params = {'n_neighbors' : [3, 5, 7, 9], 'weights' : ['uniform', 'distance'],
          'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute']}
clf = GridSearchCV(kne, params, cv = 10, scoring = 'accuracy')
clf.fit(X_train, Y_train)
print(clf.best_params_)
```

```python
knn1 = KNeighborsClassifier(algorithm = 'auto', n_neighbors = 3, weights = 'distance')
knn1.fit(X_train, Y_train)
Y_pred = knn1.predict(X_test)
print(classification_report(Y_test, Y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
params1 = {'gamma' : ['scale', 'auto'],
          'C' : [1.0, 2.0, 3.0, 4.0]}
clf = GridSearchCV(svc, params1, cv = 10, scoring = 'accuracy')
clf.fit(X_train, Y_train)
print(clf.best_params_)
```

```python
svc1 = SVC(C= 4.0, class_weight = 'balanced', gamma = 'scale')
svc1.fit(X_train, Y_train)
Y_pred = svc1.predict(X_test)
print(classification_report(Y_test, Y_pred))
```

```python
import numpy as np
nb_classifier = GaussianNB()
params2 = {'var_smoothing': np.logspace(0,-9, num=100)}
clf = GridSearchCV(gau,
                   params2,
                 cv = 10,   # use any cross validation technique
                 verbose=1,
                 scoring='accuracy')
clf.fit(X_train, Y_train)

print(clf.best_params_)
```

```python
gau1 = GaussianNB(var_smoothing = 0.012328467394420659)
gau1.fit(X_train, Y_train)
Y_pred = gau1.predict(X_test)
print(classification_report(Y_test, Y_pred))
```
